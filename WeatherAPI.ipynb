```python
import requests
import pandas as pd

def fetch_and_save_weather_data():
    # Load latitude and longitude from an Excel file
    coords_df = pd.read_excel(r'C:\Users\SalomeGrasland\Desktop\Bench\Weather Whiplash\latitude, longitude.xlsx')  # Update with your file's path

    # Define the base URL for the API call
    url = "https://archive-api.open-meteo.com/v1/archive"

    # Loop through each row in the DataFrame
    for index, row in coords_df.iterrows():
        params = {
            'latitude': row['Latitude'],  # Ensure your Excel columns are named 'latitude' and 'longitude'
            'longitude': row['Longitude'],
            'start_date': '1940-01-01',
            'end_date': '1940-12-31',
            'daily': 'temperature_2m_mean',
            'temperature_unit': 'fahrenheit'
        }

        # Make the API request
        response = requests.get(url, params=params)

        # Check if the request was successful
        if response.status_code == 200:
            data = response.json()
            # Check if the necessary data is in the response
            if 'daily' in data and 'time' in data['daily'] and 'temperature_2m_mean' in data['daily']:
                daily_data = {
                    'date': pd.to_datetime(data['daily']['time']),
                    'temperature_2m_mean': data['daily']['temperature_2m_mean']
                }
                daily_dataframe = pd.DataFrame(daily_data)
                # Generate filename using latitude and longitude
                filename = f"{params['latitude']},{params['longitude']}_weather_data.xlsx"
                # Save the DataFrame to an Excel file
                daily_dataframe.to_excel(filename, index=False)
                print(f"Data has been successfully saved to '{filename}'.")
            else:
                print("Expected data not found in the API response.")
        else:
            print(f"Failed to retrieve data for coordinates ({params['latitude']}, {params['longitude']}). Status code: {response.status_code}")

# Call the function to execute the API call and save the data
fetch_and_save_weather_data()
```
